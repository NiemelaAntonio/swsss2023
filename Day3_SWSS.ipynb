```python
import datetime
def Read_Ascii_DST(Path_to_files,Empty_Lines):
    '''Function for reading ASCII files for DST
    ____________________________________________
    
    Input Variables: 
        type(Path_to_files) = <string>  --> Path to the file that is needed to read
        type(Empty_Lines)   = <int>     --> Amount of lines that are reundant at the beginning of the file
    ____________________________________________
    
    Output Variables: 
        Date  = {'Date':[List, of, dates, datetime], 'Index':[List,of,index,values]}  --> Dictionary with 'Date' and 'Index'  
    ____________________________________________
    
    The format of the header should be  [Year  Day  Hour  Minute  SYM-H]
    
    #TODO : this can be automated so that it asserts the length of the lines to be 5, if not ignore.
    
    '''
    Data ={'dates' :[],
           'index':[]}
    with open(Path_to_files) as File:
        for i in range(Empty_Lines):
            print(File.readline())
        header        = File.readline()
        #print('this is header= '+str(header))
        variables     = header.split()
        print('these are the variables= '+str(variables))
        header        = File.readline()
        variables     = header.split()
        for line in File:
            tmp  = line.split()
            Year   =  tmp[0]
            DOY    =  tmp[1]
            Hour   =  tmp[2]
            Min    =  tmp[3]
            Date   =  datetime.datetime.strptime(str(Year)+str('-')+str(DOY)+str('-')+str(Hour)+str('-')+str(Min),'%Y-%j-%H-%M')
            #print(Date)
            Data['dates'].append(Date)
            Data['index'].append(int(tmp[4]))
        #print(Data)
        return Data
def Read_Ascii_DST_dates(Path_to_files,Empty_Lines,start_time,end_time):
    '''Function for reading ASCII files for DST
    ____________________________________________
    
    Input Variables: 
        type(Path_to_files) = <string>  --> Path to the file that is needed to read
        type(Empty_Lines)   = <int>     --> Amount of lines that are reundant at the beginning of the file
        type(start_time)    = <datetime>--> From where we want the data to start saving in the dictionary
        type(end_time)      = <datetime>--> From where we want the data to stop saving in the dictionary
    ____________________________________________
    
    Output Variables: 
        Date  = {'Date':[List, of, dates, datetime], 'Index':[List,of,index,values]}  --> Dictionary with 'Date' and 'Index'  
    ____________________________________________
    
    The format of the header should be  [Year  Day  Hour  Minute  SYM-H]
    
    #TODO : this can be automated so that it asserts the length of the lines to be 5, if not ignore.
    
    '''
    Data ={'dates' :[],
           'index':[]}
    with open(Path_to_files) as File:
        for i in range(Empty_Lines):
            File.readline()
        header        = File.readline()
        #print('this is header= '+str(header))
        variables     = header.split()
        print('these are the variables= '+str(variables))
        header        = File.readline()
        variables     = header.split()
        for line in File:
            tmp  = line.split()
            Year   =  tmp[0]
            DOY    =  tmp[1]
            Hour   =  tmp[2]
            Min    =  tmp[3]
            Date   =  datetime.datetime.strptime(str(Year)+str('-')+str(DOY)+str('-')+str(Hour)+str('-')+str(Min),'%Y-%j-%H-%M')
            #print('date before if = '+str(Date))
            #print(Date>start_time)
            if Date>start_time and Date<end_time:
                #print(Date>start_time)
                #print(Date>end_time)
                #print('date after if = '+str(Date))
                Data['dates'].append(Date)
                Data['index'].append(int(tmp[4]))
                continue
                
        #print(Data)
        return Data

if __name__ == "__main__":
    Path_to_files = "C:/Users/u0142106/Desktop/SpaceWeatherSummerSchool/swsss2023/day_02/omni_test.lst"
    Empty_lines   = 8
    Start_time    = datetime.datetime(2013,3,16,0,3)
    End_time      = datetime.datetime(2013,3,16,0,7)
    print(Read_Ascii_DST_dates(Path_to_files,
                         Empty_lines,
                         Start_time,
                         End_time))
```
