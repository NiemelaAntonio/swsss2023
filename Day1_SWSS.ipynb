Taylor approximation to a cosine function (done by us)

```python
from math import factorial
from math import pi
import numpy as np

def cos_approx(x,accuracy = 10):
    """The cosine approximation given the accuracy level.
    Generates an empty aux array
    Then calculates each element of the sumation
    Returns the sum of the elements in the aux list
    """
    aux = np.zeros(accuracy)
    for n in range(accuracy):
        aux[n] = ((-1)**n)/(factorial(2*n))*x**(2*n)
    return sum(aux)    
    
print(cos_approx(pi))
```

```python
def cos_approx_oneLine(x,accuracy = 10):
    """The cosine approximation given the accuracy level.
    This is done with list comprehension
    It calculates each element of the sumation with the list comprehension
    Returns the sum of the elements in the aux list generated beforehand
    
    Examples:
    GIVE EXAMPLES 
    """
    aux = [(((-1)**n)/(factorial(2*n))*x**(2*n)) for n in range(accuracy)]
    return sum(aux)
print(cos_approx_oneLine(pi))

def cos_approx_numpy(x,accuracy = 10):
    """The cosine approximation given the accuracy level.
    Generates an empty aux array
    Then calculates each element of the sumation
    Returns the sum of the elements in the aux list
    """
    aux = np.zeros(accuracy)
    for n in range(accuracy):
        coeff =  ((-1)**n)/(factorial(2*n))
        aux[n] =coeff*x**(2*n)
    return sum(aux)  
print(cos_approx_numpy(pi))
def cos_approx_list(x,accuracy = 10):
    """The cosine approximation given the accuracy level.
    Generates an empty aux array
    Then calculates each element of the sumation
    Returns the sum of the elements in the aux list
    """
    aux = []
    for n in range(accuracy):
        coeff =  ((-1)**n)/(factorial(2*n))
        aux.append(coeff*x**(2*n))
    return sum(aux) 
print(cos_approx_list(pi))
```

```python
def Sph_to_Cart(r,theta, phi):
    """ This should be a function that converts spherical coordinates to cartesian coordinates
    IMPORTANT! Spherical should be in degrees NOT radians
    Input variables : Spherical coordinates (r, theta , phi)
    Output          : Cartisian conversion  (x , y , z)
    
    Examples: 
    Sph_to_Cart(1,0,0) = (0.0, 0.0, 1.0)
        So the point at (r =1 , theta = 0deg , phi = 0deg) is at (x = 0, y = 0, z = 0)
        
    Sph_to_Cart(1,0,90) = (1.0, 0.0, 0.0)
        So the point at (r =1 , theta = 0deg , phi = 0deg) is at (x = 0, y = 0, z = 0)
    """ 
    x = r * np.sin(np.radians(phi)) * np.cos(np.radians(theta))
    y = r * np.sin(np.radians(phi)) * np.sin(np.radians(theta))
    z = r * np.cos(np.radians(phi))
    return (x,y,z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

r = np.linspace(-1,1)
theta = np.linspace(0,360)
phi = np.linspace(0,360)
x,y,z = Sph_to_Cart(r,theta,phi)
plt.plot(x,y,z)
plt.xlabel('x')
plt.ylabel('y')

```
